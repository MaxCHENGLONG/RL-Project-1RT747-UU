# target_update_frequency = 1000

```python
import numpy as np
import matplotlib.pyplot as plt

# .npypickle
data = np.load('loss1000.npy', allow_pickle=True)

# 
if data.dtype.kind in {'O', 'U'}:  # 
    # None
    try:
        data = np.array(data, dtype=np.float32)
    except ValueError:
        print("Data conversion error: Data might contain non-numeric values.")
        data = np.nan_to_num(data, nan=0.0)  # NoneNaN0

# 
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

smoothed_data = moving_average(data, 1000)

# 
fig, ax = plt.subplots(figsize=(15, 8))

# 
ax.plot(smoothed_data, label='SmoothL1Loss', linewidth=1, color='blue')
#ax.set_title('Loss curve when training DQN for Pong (Target_frequency = 1000)')
ax.set_xlabel('Steps')
ax.set_ylabel('SmoothL1Loss')
#ax.grid(True)

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# .npy
data = np.load('return1000.npy')  # 

# 
fig, ax = plt.subplots(figsize=(15, 8))  # 

# 
episodes = np.arange(0, 25 * len(data) , 25)  # 2525
ax.plot(episodes, data, marker='o', linestyle='-', color='blue')
#ax.set_title('Return for one round of game-play when evaluated for every 25 episodes')
ax.set_xlabel('Episode')
ax.set_ylabel('Return')
ax.grid(True)  # 

# X
ax.set_xticks(episodes)  # 25episodes

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
data = np.load('episode_length1000.npy')  # Update the path to the actual file location

# Calculate moving average with a window size of 50 (you can adjust this value)
window_size = 10
moving_avg = np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Create a figure and an axis object
fig, ax = plt.subplots(figsize=(15, 8))  # Adjust figure size to match your needs

# Plot original data
ax.plot(data, linestyle='-', color='blue', alpha=0.5, label='Original Data')  # Semi-transparent original data

# Plot moving average data
ax.plot(range(window_size-1, len(moving_avg)+window_size-1), moving_avg, linestyle='-', color='red', label='Moving Average')

# Set plot titles and labels
ax.set_xlabel('Episode')
ax.set_ylabel('Steps per Episode')
#ax.set_title('Steps per Episode with Moving Average')  # Optional: adding a title
#ax.grid(True)  # Show grid
ax.legend()  # Show legend to differentiate lines

# Display the plot
plt.show()
```
